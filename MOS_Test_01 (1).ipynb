```python
URL = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=101&oid=001&aid=0011575066'
response = requests.get(URL)
```

```python
#Status code
response.status_code
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
keyword=soup.find('div', class_='_article_body_contents')
```

```python
keyword
```

```python
print (keyword.text)
```

```python
keyword = keyword.text
```

```python
keyword = re.sub(r'[^\w]', ' ', keyword)
keyword
```

```python
keyword = re.sub('[_//^\n^\t!@#$=.%?]', '', keyword)
keyword
```

```python
#Using regex to indentify numbers and alphabets (To separate English words from Korean words)
unwanted = re.sub('[^A-Za-z0-9]+', '', keyword)
unwanted 
```

```python
for i in unwanted : 
    keyword = keyword.replace(i, '') 
```

```python
#Final refined korean words | sentences
keyword
```
